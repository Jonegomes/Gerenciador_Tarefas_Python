Etapa 3.1 -Step Estrtura de Dados

```python
import json

tarefas = []

def salvar_tarefas():
  with open("tarefas.json", "w") as arquivo:
    json.dump(tarefas, arquivo)
  print("Tarefas salvas com sucesso!")

def carregar_tarefas():
  try:
    with open("tarefas.json", 'r') as arquivo:
      tarefas_carregadas = json.load(arquivo)
      print("Tarefas carregadas com sucesso!")
      return tarefas_carregadas
  except:
    print("Nenhum arquivo de tarefas encontrado! Comecando uma nova lista.")
    return[]

def adicionar_tarefa(nome, prioridade):
  tarefa = {
        'nome': nome,
        'prioridade': prioridade
    }

  tarefas.append(tarefa)
  print("Tarefa adicionada: ", tarefa)

def listar_tarefas():
  print("LISTA DE TAREFA")
  for tarefa in tarefas:
    print(f"- {tarefa['nome']} (Prioridade {tarefa['prioridade']})")

def remover_tarefas(nome):
  for tarefa in tarefas: # Corrected: iterate through tarefas list
    if tarefa['nome'] == nome:
      tarefas.remove(tarefa)
      print("Tarefa removida!:", nome)
      break
  else:
    print("Tarefa não encontrada")

tarefas = carregar_tarefas() # Corrected: call the function

while True:
  print("** Menu de Opções **")
  print("1. Adicionar Tarefas")
  print("2. Lista de tarefas")
  print("3. Remover Tarefa")
  print("4. Salvar Tarefas")
  print("5. Sair")

  opcao = input("escolha uma Opção: ")
  if opcao == '1':
    nome = input("Digite o nome tarefa: ")
    prioridade = input("Digite a prioridade da tarefa: ")

    adicionar_tarefa(nome, prioridade)

  elif opcao == '2':
   listar_tarefas()

  elif opcao == '3':
    nome = input("Digite o nome tarefa a ser removida: ")
    remover_tarefas(nome)

  elif opcao == '4':
    salvar_tarefas()

  elif opcao == '5':
    print("saindo...")
    break

  else:
   print("Opção inválida!")
```
